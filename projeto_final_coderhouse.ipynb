```python
#importação de pacotes a serem utilizados
import pandas as pd
import requests
from plyer import notification
from datetime import datetime
```

PRIMEIRA TABELA

```python
def alerta(nivel, base, etapa):
    '''
        Alerta de falha de carregamento de base de dados
    '''
    now = str(datetime.now())

    msg = f"Falha no carregamento da base {base} na etapa {etapa}.\n{now}"

    if nivel == 1:
        title = 'ATENÇÃO: Erro Grave'
    elif nivel == 2:
        title = 'ATENÇÃO: Erro Médio'
    else:
        title = 'ATENÇÃO: Erro Baixo'
    
    notification.notify(
            title=title,
            message=msg,
            app_name='alerta',
            timeout=10
    )
```

```python
url = "https://brasilapi.com.br/api/banks/v1"
response_bancos = requests.get(url)
```

```python
status_code = response_bancos.status_code
```

```python
if status_code >= 400 and status_code < 500:
    nivel = 1
elif status_code >=500 and status_code < 600:
    nivel = 2
else:
    nivel = 3
```

```python
if response_bancos.status_code == 200:
    data_json = response_bancos.json()
else:
    alerta(nivel = nivel,
        base = "Data_Banks", 
        etapa = "EXTRACAO")
```

```python
data_banks = response_bancos.json()
df_banco = pd.DataFrame(data_banks)
```
